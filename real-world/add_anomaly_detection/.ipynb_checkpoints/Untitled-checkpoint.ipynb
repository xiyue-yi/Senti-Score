```python
from anom_detect import anom_detect
import pandas as pd
```

```python
df = pd.read_csv('score_q.txt',sep='\t',header=None)
df.head()
```

```python
pred_dict = {'time':df[0],
          'sunspots':df[2]}
pred_df = pd.DataFrame(pred_dict)
pred_df.head()
```

```python
df.drop(df.columns[0], axis=1, inplace=True)
df.index.name = 'time'
df.columns = ['sunspots']
df.head()
```

```python
an = anom_detect()
an.evaluate(pred_df)
```

```python
an.plot(df)
```

```python
an.results.head()
```
